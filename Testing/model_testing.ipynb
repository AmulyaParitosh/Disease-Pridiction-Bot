```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('../model.pkl','rb') as file:
	model = pickle.load(file)
```

```python
df = pd.read_csv('../symptoms-weight.csv')
df.set_index('Disease',inplace=True)
df.head()
```

```python
dis_df=pd.read_csv('../data.csv',index_col='disease')
dis_df.head()
```

```python
n=450

t1 = df.iloc[n,1:]

print(dis_df.loc[t1.name].name)

t1=pd.DataFrame(t1)
t1=t1.transpose()
t1
```

```python
lst=t1.iloc[0].to_list()
lst=pd.DataFrame(lst).transpose()
col = df.iloc[:,1:].columns.tolist()
lst.columns=col
lst
```

```python
p1 = model.predict(lst)[0]
p1
```

```python
out=dis_df.loc[p1]
out
```

```python
print(p1.capitalize(), "-\n", out['discription'])

precautions = out['precautions'].split(",")

print("\nPrecautions", "-")
for i,preco in enumerate(precautions):
	print(i,">",preco.lstrip())
```

```python
p_proba = model.predict_proba(lst)
p_proba
```

```python
top_n_category_predictions = np.argsort(p_proba)[:, :-n-1:-1]
order = list(top_n_category_predictions[0])[:5]
```

```python
dis_series = pd.Series(dis_df.index)
order = dis_series.loc[order].values.tolist()
```
