```python
import pickle
import pandas as pd
```

```python
with open('../model.pkl','rb') as file:
	model = pickle.load(file)
```

```python
df = pd.read_csv('../symptoms-weight.csv')
df.set_index('Disease',inplace=True)
df.head()
```

```python
dis_df=pd.read_csv('../data.csv',index_col='disease')
dis_df.head()
```

```python
n=450

t1 = df.iloc[n]

print(dis_df.loc[t1.name].name)

t1=pd.DataFrame(t1)
t1=t1.transpose()
```

```python
p1 = model.predict(t1)[0]
p1
```

```python
out=dis_df.loc[p1]
out
```

```python
print(p1.capitalize(), "-\n", out['discription'])

precautions = out['precautions'].split(",")

print("\nPrecautions", "-")
for i,preco in enumerate(precautions):
	print(i,">",preco.lstrip())
```

```python
error=0

for i in range(df.shape[0]):
	t1 = df.iloc[i]

	actual = t1.name

	t1 = pd.DataFrame(t1)
	t1 = t1.transpose()

	predicted = model.predict(t1)[0]

	if(actual!=predicted):
		error+=1

print(error)
```
