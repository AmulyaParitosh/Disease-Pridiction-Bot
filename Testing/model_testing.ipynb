```python
import pickle
import pandas as pd
```

```python
with open('../model.pkl','rb') as file:
	model = pickle.load(file)
```

```python
df = pd.read_csv('../symptoms-weight.csv')
df.rename(columns={'Unnamed: 0':"Symptom"},inplace=True)
df.set_index('Symptom',inplace=True)
```

```python
df.head()
```

```python
t1 = df.iloc[6]
t1=pd.DataFrame(t1)
t1=t1.transpose()
t1
```

```python
p1 = model.predict(t1)[0]
p1
```

```python
dis_df=pd.read_csv('../data.csv',index_col='disease')
```

```python
out=dis_df.loc[p1]
out
```

```python
print(p1.capitalize(), "-\n", out['discription'])

precautions = out['precautions'].replace('[', '')
precautions = precautions.replace(']', '')
precautions = precautions.replace("'", '')
precautions = precautions.split(",")

print("\nPrecautions", "-")
for i,preco in enumerate(precautions):
	print(i,">",preco.lstrip())
```
